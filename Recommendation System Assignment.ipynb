```python
#Data transformation

#Decide on the features that will be used for computing similarity
features = ['genre', 'rating', 'members']

# Convert categorical features into numerical representations if necessary
genre_to_id = {genre: i for i, genre in enumerate(df['genre'].unique())}
df['genre_id'] = df['genre'].map(genre_to_id)

# Normalize numerical features if required
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[['rating', 'members']] = scaler.fit_transform(df[['rating', 'members']])
```

```python
user_df = df.pivot(index='anime_id',
                                 columns='genre',
                                 values='rating')
```

```python
user_df
```

```python
#Impute those NaNs with 0 values
user_df.fillna(0, inplace=True)
user_df
```

```python
from sklearn.metrics import pairwise_distances
user_sim = 1 - pairwise_distances( user_df.values,metric='cosine')
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
# Define a function to recommend anime based on cosine similarity
def recommend_anime(target_anime_id, threshold=0.5):
    similar_anime_ids = user_sim_df.loc[target_anime_id].sort_values(ascending=False)
    similar_anime_ids = similar_anime_ids[similar_anime_ids > threshold].index
    return similar_anime_ids

# Split the dataset into training and testing sets
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Evaluate the recommendation system
def evaluate_recommendation_system(threshold=0.5):
    precision_sum = 0
    recall_sum = 0
    f1_sum = 0
    for target_anime_id in test_df['anime_id']:
        recommended_anime_ids = recommend_anime(target_anime_id, threshold)
        actual_anime_ids = test_df[test_df['anime_id'] != target_anime_id]['anime_id'].values
        true_positives = np.intersect1d(actual_anime_ids, recommended_anime_ids).size
        false_positives = recommended_anime_ids.size - true_positives
        false_negatives = actual_anime_ids.size - true_positives
        precision = true_positives / (true_positives + false_positives) if (true_positives + false_positives) > 0 else 0
        recall = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) > 0 else 0
        f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
        precision_sum += precision
        recall_sum += recall
        f1_sum += f1
    precision_avg = precision_sum / len(test_df)
    recall_avg = recall_sum / len(test_df)
    f1_avg = f1_sum / len(test_df)
    return precision_avg, recall_avg, f1_avg

# Experiment with different threshold values
thresholds = [0.3, 0.4, 0.5, 0.6, 0.7]
for threshold in thresholds:
    precision, recall, f1 = evaluate_recommendation_system(threshold)
    print(f"Threshold: {threshold:.2f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1: {f1:.4f}")
```

```python
#Interview Questions:
#1. Can you explain the difference between user-based and item-based collaborative filtering?

#Collaborative Filtering (CF) is a technique used in recommendation systems where the system makes recommendations based on past user behavior and preferences. There are two main types of collaborative filtering:

#User-Based Collaborative Filtering

#Concept: This approach recommends items to a user based on the preferences of similar users. It assumes that if users have similar tastes in the past, they will continue to have similar tastes in the future.

#Simple to understand and implement.

#Can be effective when there is a lot of user data.

#Can be less effective when users’ preferences are not diverse.

#May suffer from scalability issues with a large number of users.

#Struggles with "cold start" problems for new users who have little or no data.

#Item-Based Collaborative Filtering

#Concept: This approach recommends items based on the similarity between items. It assumes that if a user likes a certain item, they will also like similar items.

#Often more scalable than user-based methods because item similarity can be precomputed and does not change frequently.

#Better at handling new users since recommendations are based on item similarity rather than user preferences.

#More stable as item characteristics tend to be more static compared to user preferences.

#Less effective if items are very diverse or if item interactions are sparse.

#May not capture complex user preferences as effectively as user-based methods.
```

```python
#2. What is collaborative filtering, and how does it work?

#Collaborative Filtering (CF) is a method used in recommendation systems that makes predictions about a user's interests by collecting preferences or taste information from many users. The core idea is to use the preferences of many users to make recommendations based on the principle that similar users have similar tastes.

#How Collaborative Filtering Works:

#Data Collection: Gather data on user interactions with items. This can include explicit feedback like ratings or implicit feedback like clicks, purchases, or time spent on an item.

#Similarity Calculation:

#User-Based CF: Compute the similarity between users using their interaction data. Common metrics include cosine similarity, Pearson correlation, or Jaccard similarity.

#Item-Based CF: Compute the similarity between items based on how users have interacted with them.

#Recommendation Generation:

#User-Based CF: For a given user, find users who have similar preferences. Recommend items that these similar users have liked but the current user has not yet interacted with.

#Item-Based CF: For a given user, recommend items that are similar to the items they have already liked or interacted with.

#Prediction: Use the calculated similarities to predict how likely a user is to like a particular item. This is often done by averaging the ratings of similar users or items.
```
